# For Loops

```python
integers = [1,2,3,4,5]
for no in integers:
    print(no)
```

```python
for no in integers:
    print(integers)
```

```python
for no in integers:
    print('yep!')
```

```python
for i in integers:
    print(i + i)
```

```python
ice_cream_dict = {'Name' : 'Alex', 'Weekly Intake': 5,'Favorite Ice_cream':['MCC','Chocolate']}
for i in ice_cream_dict.values():
    print(i)
```

```python
for i in ice_cream_dict.keys():
    print(i)
```

```python
for keys,values in ice_cream_dict.items():
    print(keys,'->',values)
```

```python
# Nested For loop
flavors = ['Vanilla','Chocolate','Cookie Dough']
toppings = ['Hot Fudge','Oreos','Marshmallows']
for i in flavors :
    for j in toppings :
        print(i,' topped with ',j)
```

# While Lopps

```python
i = 0
while (i<5) :
    print(i)
    i = i+1
```

```python
i = 0
while (i<5):
    print(i)
    if (i == 6):
        break
    i = i+1
else:
    print('No longer < 5')
```

```python
i = 0
while (i<5):
    i = i+1
    if (i == 3):
        continue
    print(i)
else:
    print('No longer < 5')
```
