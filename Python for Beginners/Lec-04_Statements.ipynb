# If - Elif - Else Statements

```python
if (10<25) :
    print('It worked')
```

```python
if (10>25) :
    print('It worked')
else :
     print('It did not work...')
```

```python
if (10>25) :
    print('It worked')
elif (10>5) :
    print('Elif worked')
else :
     print('It did not work...')
```

```python
print('It worked') if (10>25) else print('It did not work...')
```

```python
if (10>3) :
    print('It worked')
    if (10>5) :
        print('Nested if statement worked')
elif (10>25) :
    print('Elif worked')
else :
     print('It did not work...')
```
