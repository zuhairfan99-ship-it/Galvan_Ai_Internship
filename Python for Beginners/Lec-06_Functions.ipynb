# Functions

```python
def first_func():
    print('We did it')

first_func()
```

```python
def no_squared(no):
    print(no ** 2)

no_squared(5)
```

```python
def no_squared_cust(no,pow):
    print(no ** pow)

no_squared_cust(5,3)
```

```python
no_squared_cust(pow = 5, no = 3)
```

```python
args_tuple = (5,6,1,2,8)
def no_args(*no):
    print(no[0] * no[1])

no_args(*args_tuple)
```

```python
def no_kwarg(**no):
    print('My number is : ' + no['integer'] + ' and my other number is : ' + no['integer2'])

no_kwarg(integer = '2900', integer2 = '6550')
```
