```python
# sets
daily_pints = {1,2,3,3,4,6,6,8}
wifes_daily_pints_logs = {7,3,5,7,2,1,3,4}
type(daily_pints)
```

```python
print(daily_pints)
```

```python
print(wifes_daily_pints_logs)
```

```python
print(daily_pints | wifes_daily_pints_logs )
```

```python
print(daily_pints & wifes_daily_pints_logs )
```

```python
print(daily_pints - wifes_daily_pints_logs )
```

```python
print(wifes_daily_pints_logs - daily_pints)
```

```python
print(daily_pints ^ wifes_daily_pints_logs )
```
