# Logical Operator

```python
('Vanilla' > 'Chocolate') or (50>10)
```

```python
not (50>10)
```

# Membership Operator

```python
# Membership Operator(in , notin)
ice_cream = 'I love ice_cream'
'love'in ice_cream  # -> 'love'in 'I love ice_cream'
```

```python
scoops = [1,2,3,4,5]
wanted_scoops = 8
wanted_scoops in scoops
```
