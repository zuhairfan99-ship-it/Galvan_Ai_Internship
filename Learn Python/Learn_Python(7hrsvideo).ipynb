# Continue

```python
# Whenever we want to skip any iterator in the loop, we use continue keyword. By using continue, loop will skip the statements for the iterator and move to next iterator.

for i in range(1,5):
    if i == 3:
        print ("we are printing 3")
    print (i)
```

```python
for i in range(1,5):
    if i == 3:
        print ("we are skipping 3")
        continue
    print (i)
```

```python
l1 = [10,20,30,30,20,50]
for i in l1:
    if i == 30:
        print ("skipping 30")
        continue
    print (i)
```

```python
l1 = [10,30,20,30,30,20,50]
for i in l1:
    if i == 30:
        print ("Skipping 30")
        continue 
    if i == 2 :
        print("Breaking the loop")
        break
    print (i)
```

# Recursive Function
A recursive function is defined as a function that calls itself directly or indirectly.

Functions needs to be defined first before calling

```python
# Factorial of a number

def factorial(x):
    if x == 1:
        return 1
    else:
        return x*factorial(x-1)
```

```python
factorial(5)
```

```python
factorial(6)
```

```python
factorial(7)
```

```python
# Fibinacci series
# 0 1 1 2 3 5 8 13 ...

def fibonacci(x):
    if (x == 1):
        return 0
    elif (x == 2):
        return 1 
    else :
        return (fibonacci(x-1) + fibonacci(x-2))
```

```python
fibonacci(1)
```

```python
fibonacci(2)
```

```python
fibonacci(3)
```

```python
x = int(input("Enter No. for fibonacci series:"))
l1 = []
for i in range(1, x+1):
    l1.append(fibonacci(i))
str1 = ""
for i in l1:
    str1 = str1 +" "+ str(i)
print (str1)
```

# Exception Handling

Whenever an error or exception occurs during code execution, python stops and throws an error message, due to which the code does not work further.

Avoiding such exceptions is known as Exception Handling.

In python , we can do it using try block

# try-except block

x is a variable which is not yet defined.

If we will try to print x, it will show an error, and code execution will stop.

Instead what we can do, we can provide x in try-except block.

```python
try:
    print (x)
except :
    print ("Please define x")
```

```python
x = '10.6'

try:
    y = float(x)
    print (y)
except:
    print ("Enter Number")
```

```python
x = "Hello"

try:
    y = float(x)
    print (y)
except:
    print ("Enter Number")
finally:
    print ("Finished .... ")
```

```python
def age():
    try:
        x = input("Enter Age:")
        y = int(x) + 25
        print (y)
    except:
        print ("Enter Number")
        age()
    finally:
        print ("Finished")

age()
```

# Type Error
Type error is an error which occurs when two variables of different datatype perform an operation.

```python
x =2
y = 'Hi'
try:
    print (x + y)
except TypeError:
    print ("Only integers allowed")
finally:
    print ("Finished ... ")
```

# Value Error
Value error is an error which occurs when a variable receives an argument of correct type but improper value.

```python
x = 'hello'
try:
    print (int(x))
except ValueError:    
    print ("Only numbers allowed")
finally:
    print ("Finished ... ")
```

# ZeroDivisionError
Zero Division error is an error which occurs when a number is divided by o.

```python
x = 2
y = 0
try:
    print (x/y)
except ZeroDivisionError:
    print ("Division by 0 is not allowed.")
finally:
    print ("Finished ... ")    
```

# Finnding eLement using index
In order to access any array element by using its position is called as indexing.

We can find the element by using its index position as 1st position 

`arr = np.array([10,20,30])`

`arr[0]` -> Will return 10 as 10 is at 0th position or 0th index

`arr[2]` -> Will return 30 as 30 is at 2nd position or 2nd index

# Finding element using negative index
In order to access elements from end, we use negative indexing.

Index :- I is considered as last element of array followed by -2 as 2nd last and so on.

`arr = np.array([10,20,30])`

`arr[-1]` -> Will return 30 as 30 is at -1th position or -1th index

# Slicing elements from array
Slicing means accessing elements from one index position to another.

`arr = np.array([10,20,30,40,50,60,70])`

`10 20 30 40 50 60 70` -> `10 20 30`
 -> `30 40 50 60 70`       -> `20 30 40 50`
  

# Slicing Elements From Array
Slicing can be done as arr[start : stop] where slice begins from start index and ends at (stop-1).

`arr[start : stop]` -> Slicing can be done as arr[start : stop : step] where steps can also be added.

`arr[start : stop : step]`

If start is not provided, it will be considered as 0.

If stop is not provided it will be considered as length of array +I.

If step is not provided, it will be considered as I.

# Filtering elements using positions as list
If instead of slicing from 0-n or I-n, we want to filter elements from random positions say (1,2,4,7), we can do so by providing list of such values as argument.

`ListI = [1,2,4,7]`

`arr[List I]` -> Will filter 1,2,4,7th index element


```python
import numpy as np

arr1 = np.array([10,20,30,40,50,60,70])
arr1
```
