# Project 5: Task Manager with Priority System

## Import Required Libraries

```python
import pandas as pd
```

## Step 1: Create the Data Storage

```python
tasks = []
next_task_id = 1
```

## Step 2: Function to Add a Task

```python
def add_task(tsk_n, assg_to, pty, due_d):
    global next_task_id
    tsk = {
        "task_id": next_task_id,
        "task_name": tsk_n,
        "assigned_to": assg_to,
        "priority": pty,
        "status": "Pending",
        "due_date": due_d
    }
    tasks.append(tsk)
    next_task_id = next_task_id + 1
    print("Task added:", tsk_n)
```

## Step 3: Function to View All Tasks

```python
def view_tasks():
    for tsk in tasks:
        print(tsk["task_id"], "-", tsk["task_name"], "-", tsk["assigned_to"], "-", tsk["priority"], "-", tsk["status"], "-", tsk["due_date"])
```

## Step 4: Function to Update Task Status

```python
def update_task_status(tsk_id, new_st):
    for tsk in tasks:
        if tsk["task_id"] == tsk_id:
            tsk["status"] = new_st
            print("Task", tsk_id, "status updated to", new_st)
            return
    print("Task", tsk_id, "not found")
```

## Step 5: Function to Filter Tasks by Status

```python
def filter_tasks_by_status(st):
    filtered_list = []
    for tsk in tasks:
        if tsk["status"] == st:
            filtered_list.append(tsk)
    return filtered_list
```

## Step 6: Function to Filter Tasks by Priority

```python
def filter_tasks_by_priority(p):
    filtered_list = []
    for tsk in tasks:
        if tsk["priority"] == p:
            filtered_list.append(tsk)
    return filtered_list
```

## Step 7: Function to Count Tasks by Status

```python
def count_tasks_by_status():
    st_c = {}
    for tsk in tasks:
        st = tsk["status"]
        if st in st_c:
            st_c[st] = st_c[st] + 1
        else:
            st_c[st] = 1
    return st_c
```

## Step 8: Function to Count Tasks by Assignee

```python
def count_tasks_by_assignee():
    assig_c = {}
    for tsk in tasks:
        assig = tsk["assigned_to"]
        if assig in assig_c:
            assig_c[assig] = assig_c[assig] + 1
        else:
            assig_c[assig] = 1
    return assig_c
```

## Step 9: Function to Display Task Report Using Pandas

```python
def display_task_report():
    df = pd.DataFrame(tasks)
    print(df)
    return df
```

## Step 10: Adding Sample Tasks

```python
add_task("Design homepage", "Zuha", "High", "2026-07-15")
add_task("Write blog post", "Sara", "Medium", "2026-07-18")
add_task("Fix login bug", "Bilal", "High", "2026-07-12")
add_task("Update documentation", "Sara", "Low", "2026-07-20")
add_task("Test payment gateway", "Ali", "Medium", "2026-07-16")
```

## Step 11: Viewing All Tasks

```python
view_tasks()
```

## Step 12: Updating a Task Status

```python
update_task_status(3, "In Progress")
update_task_status(1, "Completed")
```

## Step 13: Viewing Pending Tasks

```python
pend_tsks = filter_tasks_by_status("Pending")
for tsk in pend_tsks:
    print(tsk["task_name"], "assigned to", tsk["assigned_to"])
```

## Step 14: Viewing Completed Tasks

```python
cmp_tsk = filter_tasks_by_status("Completed")
for tsk in cmp_tsk:
    print(tsk["task_name"], "assigned to", tsk["assigned_to"])
```

## Step 15: Viewing High Priority Tasks

```python
high_pri_tsk = filter_tasks_by_priority("High")
for tsk in high_pri_tsk:
    print(tsk["task_name"], "assigned to", tsk["assigned_to"])
```

## Step 16: Counting Tasks by Status

```python
st_c = count_tasks_by_status()
print(st_c)
```

## Step 17: Counting Tasks by Assignee

```python
assig_c = count_tasks_by_assignee()
print(assig_c)
```

## Step 18: Displaying the Task Report Using Pandas

```python
display_task_report()
```

## Step 19: Challenge - Menu Based System

```python
def run_menu():
    while True:
        print("1. Add task")
        print("2. View all tasks")
        print("3. Update task status")
        print("4. View pending tasks")
        print("5. View completed tasks")
        print("6. View high priority tasks")
        print("7. Count tasks by status")
        print("8. Count tasks by assignee")
        print("9. Display task report")
        print("10. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            tsk_n = input("Enter task name: ")
            assig_to = input("Enter assignee name: ")
            pri = input("Enter priority (High/Medium/Low): ")
            due_d = input("Enter due date: ")
            add_task(tsk_n, assig_to, pri, due_d)
        elif choice == "2":
            view_tasks()
        elif choice == "3":
            tsk_id = int(input("Enter task ID: "))
            new_st = input("Enter new status (Pending/In Progress/Completed/Blocked): ")
            update_task_status(tsk_id, new_st)
        elif choice == "4":
            pend_tsk = filter_tasks_by_status("Pending")
            for tsk in pend_tsk:
                print(tsk["task_name"], "assigned to", tsk["assigned_to"])
        elif choice == "5":
            cmp_tsk = filter_tasks_by_status("Completed")
            for tsk in cmp_tsk:
                print(tsk["task_name"], "assigned to", tsk["assigned_to"])
        elif choice == "6":
            high_pri_tsk = filter_tasks_by_priority("High")
            for tsk in high_pri_tsk:
                print(tsk["task_name"], "assigned to", tsk["assigned_to"])
        elif choice == "7":
            st_c = count_tasks_by_status()
            print(st_c)
        elif choice == "8":
            assig_c = count_tasks_by_assignee()
            print(assig_c)
        elif choice == "9":
            display_task_report()
        elif choice == "10":
            print("Exiting the program")
            break
        else:
            print("Invalid choice, please try again")
```

```python
run_menu()
```
