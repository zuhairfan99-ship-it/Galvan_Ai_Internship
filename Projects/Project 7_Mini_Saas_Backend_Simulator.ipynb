# Mini SaaS Backend Simulator
## AI Automation Agency Operations System

```python
import pandas as pd   
```

```python
leads = []                    
clients = []
projects = []
tasks = []
tickets = []
next_lead_id = 1
next_client_id = 1                 
next_project_id = 1           
next_task_id = 1
next_ticket_id = 1
################################################################################################################################################
def generate_id(curr_id):
    return curr_id + 1           
```

## Module 1: Lead Management

```python
def calculate_lead_score(bdg, u, decision_maker):
    scr = 0
    if bdg > 3000:
        scr = scr + 30
    elif bdg >= 1000:
        scr = scr + 20
    else:
        scr = scr + 10
    if u.lower() == "immediate":
        scr = scr + 30
    elif u.lower() == "this month":
        scr = scr + 20
    else:
        scr = scr + 10
    if decision_maker.lower() == "yes":
        scr = scr + 30
    else:
        scr = scr + 10
    return scr
################################################################################################################################################
def classify_lead(scr):
    if scr >= 70:
        return "Hot"
    elif scr >= 40:
        return "Warm"
    else:
        return "Cold"
################################################################################################################################################
def add_lead(n, cmp, srv_needed, bdg, u, decision_maker):
    global next_lead_id
    scr = calculate_lead_score(bdg, u, decision_maker)
    st = classify_lead(scr)
    ld = {
        "lead_id": next_lead_id,
        "name": n,
        "company": cmp,
        "service_needed": srv_needed,
        "budget": bdg,
        "urgency": u,
        "decision_maker": decision_maker,
        "lead_score": scr,
        "lead_status": st
    }
    leads.append(ld)
    next_lead_id = generate_id(next_lead_id)
    print("Lead added:", n, "-", "Score:", scr, "-", "Status:", st)
################################################################################################################################################
def view_leads():
    df = pd.DataFrame(leads)
    print(df)
    return df
################################################################################################################################################
def filter_hot_leads():
    hot_leads = []
    for ld in leads:
        if ld["lead_status"] == "Hot":
            hot_leads.append(ld)
    return hot_leads
################################################################################################################################################
def search_lead(key):
    results = []
    for ld in leads:
        if key.lower() in ld["name"].lower() or key.lower() in ld["company"].lower():
            results.append(ld)
    return results
################################################################################################################################################
def sort_leads_by_score():
    sort_ld = sorted(leads, key=lambda ld: ld["lead_score"], reverse=True)
    return sort_ld
################################################################################################################################################
def convert_lead_to_client(ld_id, email):
    for ld in leads:
        if ld["lead_id"] == ld_id and ld["lead_status"] == "Hot":
            add_client(ld["name"], ld["company"], email, ld["service_needed"])
            print(ld["name"], "has been converted from a lead into a client")
            return
    print("Lead not found or not a hot lead")
```

## Module 2: Client Management

```python
def add_client(n, cmp, email, srv_purchased):
    global next_client_id
    cl = {
        "client_id": next_client_id,
        "name": n,
        "company": cmp,
        "email": email,
        "service_purchased": srv_purchased,
        "project_id": None,
        "payment_status": "Pending"
    }
    clients.append(cl)
    next_client_id = generate_id(next_client_id)
    print("Client added:", n)
################################################################################################################################################
def view_clients():
    df = pd.DataFrame(clients)
    print(df)
    return df
#################################################################################################################################################
def search_client(cmp):
    results = []
    for c in clients:
        if cmp.lower() in c["company"].lower():
            results.append(c)
    return results
################################################################################################################################################
def update_payment_status(c_id, new_st):
    for c in clients:
        if c["client_id"] == c_id:
            c["payment_status"] = new_st
            print("Client", c_id, "payment status updated to", new_st)
            return
    print("Client not found")
################################################################################################################################################
def show_unpaid_clients():
    unpaid_clients = []
    for c in clients:
        if c["payment_status"] == "Pending" or c["payment_status"] == "Overdue":
            unpaid_clients.append(c)
    return unpaid_clients
```

## Module 3: Project Management

```python
def create_project(client_id, project_name, service_type, start_date, deadline):
    global next_project_id
    client_name = None
    for client in clients:
        if client["client_id"] == client_id:
            client_name = client["name"]
            client["project_id"] = next_project_id
    project = {
        "project_id": next_project_id,
        "client_name": client_name,
        "project_name": project_name,
        "service_type": service_type,
        "start_date": start_date,
        "deadline": deadline,
        "project_status": "Not Started"
    }
    projects.append(project)
    next_project_id = generate_id(next_project_id)
    print("Project created:", project_name, "for", client_name)
#################################################################################################################################################
def view_projects():
    data_frame = pd.DataFrame(projects)
    print(data_frame)
    return data_frame
#################################################################################################################################################
def update_project_status(project_id, new_status):
    for project in projects:
        if project["project_id"] == project_id:
            project["project_status"] = new_status
            print("Project", project_id, "status updated to", new_status)
            return
    print("Project not found")
#################################################################################################################################################
def show_projects_by_status(status):
    filtered_projects = []
    for project in projects:
        if project["project_status"] == status:
            filtered_projects.append(project)
    return filtered_projects
#################################################################################################################################################
def count_projects_by_status():
    status_count = {}
    for project in projects:
        status = project["project_status"]
        if status in status_count:
            status_count[status] = status_count[status] + 1
        else:
            status_count[status] = 1
    return status_count
#################################################################################################################################################
def sort_projects_by_deadline():
    sorted_projects = sorted(projects, key=lambda project: project["deadline"])
    return sorted_projects
```

## Module 4: Task Management

```python
def add_task(project_id, task_name, assigned_to, priority):
    global next_task_id
    task = {
        "task_id": next_task_id,
        "project_id": project_id,
        "task_name": task_name,
        "assigned_to": assigned_to,
        "priority": priority,
        "status": "Pending"
    }
    tasks.append(task)
    next_task_id = generate_id(next_task_id)
    print("Task added:", task_name)
#################################################################################################################################################
def view_tasks():
    data_frame = pd.DataFrame(tasks)
    print(data_frame)
    return data_frame
#################################################################################################################################################
def view_tasks_by_project(project_id):
    filtered_tasks = []
    for task in tasks:
        if task["project_id"] == project_id:
            filtered_tasks.append(task)
    return filtered_tasks
#################################################################################################################################################
def update_task_status(task_id, new_status):
    for task in tasks:
        if task["task_id"] == task_id:
            task["status"] = new_status
            print("Task", task_id, "status updated to", new_status)
            return
    print("Task not found")
#################################################################################################################################################
def filter_tasks_by_assignee(assigned_to):
    filtered_tasks = []
    for task in tasks:
        if task["assigned_to"] == assigned_to:
            filtered_tasks.append(task)
    return filtered_tasks
#################################################################################################################################################
def filter_tasks_by_priority(priority):
    filtered_tasks = []
    for task in tasks:
        if task["priority"] == priority:
            filtered_tasks.append(task)
    return filtered_tasks
#################################################################################################################################################
def show_high_priority_pending_tasks():
    filtered_tasks = []
    for task in tasks:
        if task["priority"] == "High" and task["status"] == "Pending":
            filtered_tasks.append(task)
    return filtered_tasks
```

## Module 5: Support Ticket Management

```python
def assign_ticket_priority(issue):
    issue_lower = issue.lower()
    if "not working" in issue_lower or "error" in issue_lower or "crash" in issue_lower:
        return "High"
    elif "payment" in issue_lower or "refund" in issue_lower or "invoice" in issue_lower:
        return "Medium"
    else:
        return "Low"
#################################################################################################################################################
def add_ticket(client_name, issue, category):
    global next_ticket_id
    priority = assign_ticket_priority(issue)
    ticket = {
        "ticket_id": next_ticket_id,
        "client_name": client_name,
        "issue": issue,
        "category": category,
        "priority": priority,
        "status": "Open"
    }
    tickets.append(ticket)
    next_ticket_id = generate_id(next_ticket_id)
    print("Ticket added for", client_name, "-", "Priority:", priority)
#################################################################################################################################################
def view_tickets():
    data_frame = pd.DataFrame(tickets)
    print(data_frame)
    return data_frame
#################################################################################################################################################
def close_ticket(ticket_id):
    for ticket in tickets:
        if ticket["ticket_id"] == ticket_id:
            ticket["status"] = "Closed"
            print("Ticket", ticket_id, "has been closed")
            return
    print("Ticket not found")
#################################################################################################################################################
def view_open_tickets():
    open_tickets = []
    for ticket in tickets:
        if ticket["status"] == "Open":
            open_tickets.append(ticket)
    return open_tickets
#################################################################################################################################################
def view_high_priority_tickets():
    high_priority_tickets = []
    for ticket in tickets:
        if ticket["priority"] == "High":
            high_priority_tickets.append(ticket)
    return high_priority_tickets
```

```python
def generate_final_report():
    total_leads = len(leads)
    total_hot_leads = len(filter_hot_leads())
    total_clients = len(clients)
    total_projects = len(projects)
    completed_projects = 0
    for project in projects:
        if project["project_status"] == "Completed":
            completed_projects = completed_projects + 1
    pending_payments = len(show_unpaid_clients())
    total_tasks = len(tasks)
    completed_tasks = 0
    for task in tasks:
        if task["status"] == "Completed":
            completed_tasks = completed_tasks + 1
    open_tickets = len(view_open_tickets())
    high_priority_tickets = len(view_high_priority_tickets())
    print("----- FINAL DASHBOARD -----")
    print("Total Leads:", total_leads)
    print("Total Hot Leads:", total_hot_leads)
    print("Total Clients:", total_clients)
    print("Total Projects:", total_projects)
    print("Completed Projects:", completed_projects)
    print("Pending Payments:", pending_payments)
    print("Total Tasks:", total_tasks)
    print("Completed Tasks:", completed_tasks)
    print("Open Support Tickets:", open_tickets)
    print("High Priority Support Tickets:", high_priority_tickets)

generate_final_report()
```

## Menu System

```python
def lead_menu():
    print("1. Add lead")
    print("2. View all leads")
    print("3. Filter hot leads")
    print("4. Search lead")
    print("5. Sort leads by score")
    print("6. Convert lead to client")
    print("7. Back to main menu")
    choice = input("Enter your choice: ")
    if choice == "1":
        name = input("Enter lead name: ")
        company = input("Enter company: ")
        service_needed = input("Enter service needed: ")
        budget = float(input("Enter budget: "))
        urgency = input("Enter urgency (Immediate/This month/Later): ")
        decision_maker = input("Is this person a decision maker? (Yes/No): ")
        add_lead(name, company, service_needed, budget, urgency, decision_maker)
    elif choice == "2":
        view_leads()
    elif choice == "3":
        hot_leads = filter_hot_leads()
        for lead in hot_leads:
            print(lead["name"], "from", lead["company"])
    elif choice == "4":
        keyword = input("Enter name or company to search: ")
        results = search_lead(keyword)
        for lead in results:
            print(lead["name"], "-", lead["company"])
    elif choice == "5":
        sorted_leads = sort_leads_by_score()
        for lead in sorted_leads:
            print(lead["name"], "-", lead["lead_score"])
    elif choice == "6":
        lead_id = int(input("Enter lead ID to convert: "))
        email = input("Enter client email: ")
        convert_lead_to_client(lead_id, email)
    elif choice == "7":
        return
    else:
        print("Invalid choice")
```

```python
def client_menu():
    print("1. Add client")
    print("2. View all clients")
    print("3. Search client by company")
    print("4. Update payment status")
    print("5. Show unpaid clients")
    print("6. Back to main menu")
    choice = input("Enter your choice: ")
    if choice == "1":
        name = input("Enter client name: ")
        company = input("Enter company: ")
        email = input("Enter email: ")
        service_purchased = input("Enter service purchased: ")
        add_client(name, company, email, service_purchased)
    elif choice == "2":
        view_clients()
    elif choice == "3":
        company = input("Enter company name to search: ")
        results = search_client(company)
        for client in results:
            print(client["name"], "-", client["company"])
    elif choice == "4":
        client_id = int(input("Enter client ID: "))
        new_status = input("Enter new payment status (Paid/Pending/Overdue): ")
        update_payment_status(client_id, new_status)
    elif choice == "5":
        unpaid_clients = show_unpaid_clients()
        for client in unpaid_clients:
            print(client["name"], "-", client["payment_status"])
    elif choice == "6":
        return
    else:
        print("Invalid choice")
```

```python
def project_menu():
    print("1. Create project")
    print("2. View all projects")
    print("3. Update project status")
    print("4. Show completed projects")
    print("5. Show delayed projects")
    print("6. Count projects by status")
    print("7. Sort projects by deadline")
    print("8. Back to main menu")
    choice = input("Enter your choice: ")
    if choice == "1":
        client_id = int(input("Enter client ID: "))
        project_name = input("Enter project name: ")
        service_type = input("Enter service type: ")
        start_date = input("Enter start date: ")
        deadline = input("Enter deadline: ")
        create_project(client_id, project_name, service_type, start_date, deadline)
    elif choice == "2":
        view_projects()
    elif choice == "3":
        project_id = int(input("Enter project ID: "))
        new_status = input("Enter new status (Not Started/In Progress/Completed/Delayed): ")
        update_project_status(project_id, new_status)
    elif choice == "4":
        completed_projects = show_projects_by_status("Completed")
        for project in completed_projects:
            print(project["project_name"], "-", project["client_name"])
    elif choice == "5":
        delayed_projects = show_projects_by_status("Delayed")
        for project in delayed_projects:
            print(project["project_name"], "-", project["client_name"])
    elif choice == "6":
        status_count = count_projects_by_status()
        print(status_count)
    elif choice == "7":
        sorted_projects = sort_projects_by_deadline()
        for project in sorted_projects:
            print(project["project_name"], "-", project["deadline"])
    elif choice == "8":
        return
    else:
        print("Invalid choice")
```

```python
def task_menu():
    print("1. Add task to project")
    print("2. View all tasks")
    print("3. View tasks by project")
    print("4. Update task status")
    print("5. View tasks by assignee")
    print("6. Show high priority pending tasks")
    print("7. Back to main menu")
    choice = input("Enter your choice: ")
    if choice == "1":
        project_id = int(input("Enter project ID: "))
        task_name = input("Enter task name: ")
        assigned_to = input("Enter assignee name: ")
        priority = input("Enter priority (High/Medium/Low): ")
        add_task(project_id, task_name, assigned_to, priority)
    elif choice == "2":
        view_tasks()
    elif choice == "3":
        project_id = int(input("Enter project ID: "))
        project_tasks = view_tasks_by_project(project_id)
        for task in project_tasks:
            print(task["task_name"], "-", task["assigned_to"])
    elif choice == "4":
        task_id = int(input("Enter task ID: "))
        new_status = input("Enter new status (Pending/In Progress/Completed/Blocked): ")
        update_task_status(task_id, new_status)
    elif choice == "5":
        assigned_to = input("Enter assignee name: ")
        assignee_tasks = filter_tasks_by_assignee(assigned_to)
        for task in assignee_tasks:
            print(task["task_name"], "-", task["priority"])
    elif choice == "6":
        high_priority_tasks = show_high_priority_pending_tasks()
        for task in high_priority_tasks:
            print(task["task_name"], "-", task["assigned_to"])
    elif choice == "7":
        return
    else:
        print("Invalid choice")
```

```python
def ticket_menu():
    print("1. Add support ticket")
    print("2. View all tickets")
    print("3. Close ticket")
    print("4. View open tickets")
    print("5. View high priority tickets")
    print("6. Back to main menu")
    choice = input("Enter your choice: ")
    if choice == "1":
        client_name = input("Enter client name: ")
        issue = input("Describe the issue: ")
        category = input("Enter category (Billing/Technical/Account/Feature Request/Other): ")
        add_ticket(client_name, issue, category)
    elif choice == "2":
        view_tickets()
    elif choice == "3":
        ticket_id = int(input("Enter ticket ID to close: "))
        close_ticket(ticket_id)
    elif choice == "4":
        open_tickets = view_open_tickets()
        for ticket in open_tickets:
            print(ticket["client_name"], "-", ticket["issue"])
    elif choice == "5":
        high_priority_tickets = view_high_priority_tickets()
        for ticket in high_priority_tickets:
            print(ticket["client_name"], "-", ticket["issue"])
    elif choice == "6":
        return
    else:
        print("Invalid choice")
```

```python
def main_menu():
    while True:
        print()
        print("========== MAIN MENU ==========")
        print("1. Lead Management")
        print("2. Client Management")
        print("3. Project Management")
        print("4. Task Management")
        print("5. Support Ticket Management")
        print("6. Final Dashboard")
        print("7. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            lead_menu()
        elif choice == "2":
            client_menu()
        elif choice == "3":
            project_menu()
        elif choice == "4":
            task_menu()
        elif choice == "5":
            ticket_menu()
        elif choice == "6":
            generate_final_report()
        elif choice == "7":
            print("Exiting the program")
            break
        else:
            print("Invalid choice, please try again")
```
