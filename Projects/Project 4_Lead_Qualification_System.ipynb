# Project 4: Lead Qualification System

## Import Required Libraries

```python
import pandas as pd
```

## Step 1: Create the Data Storage

```python
leads = []
```

## Step 2: Function to Add a Lead

```python
def add_lead(n, c, bdg, u, srv_needed, decision_maker):
    lead = {
        "name": n,
        "company": c,
        "budget": bdg,
        "urgency": u,
        "service_needed": srv_needed,
        "decision_maker": decision_maker
    }
    leads.append(lead)
    print(n, "from", c, "has been added as a lead")
```

## Step 3: Function to Calculate Budget Score

```python
def calculate_budget_score(bdg):
    if bdg > 3000:
        return 30
    elif bdg >= 1000:
        return 20
    else:
        return 10
```

## Step 4: Function to Calculate Urgency Score

```python
def calculate_urgency_score(u):
    if u.lower() == "immediate":
        return 30
    elif u.lower() == "this month":
        return 20
    else:
        return 10
```

## Step 5: Function to Calculate Decision Maker Score

```python
def calculate_decision_maker_score(decision_maker):
    if decision_maker.lower() == "yes":
        return 30
    else:
        return 10
```

## Step 6: Function to Calculate Total Score

```python
def calculate_total_score(lead):
    bdg_score = calculate_budget_score(lead["budget"])
    u_score = calculate_urgency_score(lead["urgency"])
    decision_score = calculate_decision_maker_score(lead["decision_maker"])
    t_score = bdg_score + u_score + decision_score
    return t_score
```

## Step 7: Function to Classify a Lead

```python
def classify_lead(t_score):
    if t_score >= 70:
        return "Hot Lead"
    elif t_score >= 40:
        return "Warm Lead"
    else:
        return "Cold Lead"
```

## Step 8: Function to Process All Leads

```python
def process_leads():
    for ld in leads:
        t_score = calculate_total_score(ld)
        ld["score"] = t_score
        ld["status"] = classify_lead(t_score)
```

## Step 9: Function to Display All Leads Using Pandas

```python
def display_leads():
    process_leads()
    df = pd.DataFrame(leads)
    print(df)
    return df
```

## Step 10: Function to Filter Hot Leads

```python
def filter_hot_leads():
    process_leads()
    hot_leads = []
    for ld in leads:
        if ld["status"] == "Hot Lead":
            hot_leads.append(ld)
    return hot_leads
```

## Step 11: Adding Sample Leads

```python
add_lead("Zuha Irfan", "Irfan Traders", 5000, "Immediate", "Website Development", "Yes")
add_lead("Fatima Noor", "Noor Boutique", 1500, "This month", "Social Media Marketing", "No")
add_lead("Usman Tariq", "Tariq Electronics", 800, "Later", "SEO Services", "No")
add_lead("Ayesha Farooq", "Farooq Consultants", 4200, "Immediate", "AI Automation", "Yes")
```

## Step 12: Displaying All Leads

```python
display_leads()
```

## Step 13: Filtering Hot Leads

```python
hot_leads = filter_hot_leads()
for ld in hot_leads:
    print(ld["name"], "from", ld["company"], "is a Hot Lead")
```

## Step 14: Challenge - Final Report

```python
def generate_final_report():
    process_leads()
    total_ld = len(leads)
    hot_c = 0
    warm_c = 0
    cold_c = 0
    t_score = 0
    for ld in leads:
        t_score = t_score + ld["score"]
        if ld["status"] == "Hot Lead":
            hot_c = hot_c + 1
        elif ld["status"] == "Warm Lead":
            warm_c = warm_c + 1
        else:
            cold_c = cold_c + 1
    avg_score = t_score / total_ld
    best_ld = leads[0]
    for ld in leads:
        if ld["score"] > best_ld["score"]:
            best_ld = ld
    print("Total Leads:", total_ld)
    print("Hot Leads:", hot_c)
    print("Warm Leads:", warm_c)
    print("Cold Leads:", cold_c)
    print("Average Lead Score:", avg_score)
    print("Best Lead to Contact First:", best_ld["name"], "from", best_ld["company"])
```

```python
generate_final_report()
```

## Step 15: Interactive Menu

```python
def run_menu():
    while True:
        print("1. Add lead")
        print("2. View all leads")
        print("3. View hot leads")
        print("4. View final report")
        print("5. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            n = input("Enter lead name: ")
            c = input("Enter company name: ")
            bdg = float(input("Enter budget: "))
            u = input("Enter urgency (Immediate / This month / Later): ")
            srv_needed = input("Enter service needed: ")
            decision_maker = input("Is this person a decision maker? (Yes/No): ")
            add_lead(n, c, bdg, u, srv_needed, decision_maker)
        elif choice == "2":
            display_leads()
        elif choice == "3":
            hot_ld = filter_hot_leads()
            for ld in hot_leads:
                print(ld["name"], "from", ld["company"], "is a Hot Lead")
        elif choice == "4":
            generate_final_report()
        elif choice == "5":
            print("Exiting the program")
            break
        else:
            print("Invalid choice, please try again")
```
