# Project 6: Support Ticket Analyzer

## Import Required Libraries

```python
import pandas as pd
```

## Step 1: Create the Data Storage

```python
tickets = []
next_ticket_id = 1
```

## Step 2: Function to Detect Category

```python
def detect_category(i):
    i_low = i.lower()
    if "bill" in i_low or "invoice" in i_low or "payment" in i_low or "refund" in i_low:
        return "Billing"
    elif "error" in i_low or "crash" in i_low or "not working" in i_low or "bug" in i_low:
        return "Technical"
    elif "login" in i_low or "password" in i_low or "account" in i_low:
        return "Account"
    elif "request" in i_low or "feature" in i_low or "add" in i_low:
        return "Feature Request"
    else:
        return "Other"
```

## Step 3: Function to Assign Priority

```python
def assign_priority(i):
    i_low = i.lower()
    if "not working" in i_low or "error" in i_low or "crash" in i_low:
        return "High"
    elif "payment" in i_low or "refund" in i_low or "invoice" in i_low:
        return "Medium"
    elif "how to" in i_low or "request" in i_low or "change" in i_low:
        return "Low"
    else:
        return "Medium"
```

## Step 4: Function to Add a Ticket

```python
def add_ticket(customer_n, i):
    global next_ticket_id
    c = detect_category(i)
    pri = assign_priority(i)
    tkt = {
        "ticket_id": next_ticket_id,
        "customer_name": customer_n,
        "issue": i,
        "category": c,
        "priority": pri,
        "status": "Open"
    }
    tickets.append(tkt)
    next_ticket_id = next_ticket_id + 1
    print("Ticket added for", customer_n, "-", "Category:", c, "-", "Priority:", pri)
```

## Step 5: Function to View All Tickets

```python
def view_tickets():
    for tkt in tickets:
        print(tkt["ticket_id"], "-", tkt["customer_name"], "-", tkt["category"], "-", tkt["priority"], "-", tkt["status"])
```

## Step 6: Function to Filter Tickets by Priority

```python
def filter_by_priority(pri):
    filtered_list = []
    for tkt in tickets:
        if tkt["priority"] == pri:
            filtered_list.append(tkt)
    return filtered_list
```

## Step 7: Function to Filter Tickets by Status

```python
def filter_by_status(st):
    filtered_list = []
    for tkt in tickets:
        if tkt["status"] == st:
            filtered_list.append(tkt)
    return filtered_list
```

## Step 8: Function to Close a Ticket

```python
def close_ticket(tkt_id):
    for tkt in tickets:
        if tkt["ticket_id"] == tkt_id:
            tkt["status"] = "Closed"
            print("Ticket", tkt_id, "has been closed")
            return
    print("Ticket", tkt_id, "not found")
```

## Step 9: Function to Generate Ticket Report Using Pandas

```python
def generate_ticket_report():
    df = pd.DataFrame(tickets)
    print(df)
    return df
```

## Step 16: Challenge - Final Report

```python
def generate_final_report():
    total_tkt = len(tickets)
    open_c = 0
    closed_c = 0
    high_pri_c = 0
    category_c = {}
    pri_gp = {}
    for tkt in tickets:
        if tkt["status"] == "Open":
            open_c = open_c + 1
        else:
            closed_c = closed_c + 1
        if tkt["priority"] == "High":
            high_pri_c = high_pri_c + 1
        c = tkt["category"]
        if c in category_c:
            category_c[c] = category_c[c] + 1
        else:
            category_c[c] = 1
        pri = tkt["priority"]
        if pri in pri_gp:
            pri_gp[pri] = pri_gp[pri] + 1
        else:
            pri_gp[pri] = 1
    most_common_category = None
    high_c = 0
    for c in category_c:
        if category_c[c] > high_c:
            high_c = category_c[c]
            most_common_category = c
    print("Total Tickets:", total_tkt)
    print("Open Tickets:", open_c)
    print("Closed Tickets:", closed_c)
    print("High Priority Tickets:", high_pri_c)
    print("Most Common Category:", most_common_category)
    print("Tickets Grouped by Priority:", pri_gp)
```

```python
generate_final_report()
```

## Step 17: Interactive Menu

```python
def run_menu():
    while True:
        print("1. Add ticket")
        print("2. View all tickets")
        print("3. Close ticket")
        print("4. View tickets by priority")
        print("5. View tickets by status")
        print("6. View final report")
        print("7. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            customer_n = input("Enter customer name: ")
            i = input("Describe the issue: ")
            add_ticket(customer_n, i)
        elif choice == "2":
            view_tickets()
        elif choice == "3":
            tkt_id = int(input("Enter ticket ID to close: "))
            close_ticket(tkt_id)
        elif choice == "4":
            pri = input("Enter priority (High/Medium/Low): ")
            filtered_tkt = filter_by_priority(pri)
            for tkt in filtered_tkt:
                print(tkt["customer_name"], "-", tkt["issue"])
        elif choice == "5":
            st = input("Enter status (Open/Closed): ")
            filtered_tkt = filter_by_status(st)
            for tkt in filtered_tkt:
                print(tkt["customer_name"], "-", tkt["issue"])
        elif choice == "6":
            generate_final_report()
        elif choice == "7":
            print("Exiting the program")
            break
        else:
            print("Invalid choice, please try again")
```

```python
run_menu()
```
