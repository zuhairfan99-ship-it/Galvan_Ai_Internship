# Project 2: Expense Tracker

## Import Required Libraries

```python
import pandas as pd
```

## Step 1: Create the Data Storage

```python
expenses = []
```

## Step 2: Function to Add an Expense

```python
def add_expense(c, a, d):
    exp = {
        "category": c,
        "amount": a,
        "date": d
    }
    expenses.append(exp)
    print("Expense added:", c, a, d)
```

## Step 3: Function to View All Expenses

```python
def view_expenses():
    for exp in expenses:
        print(exp["date"], "-", exp["category"], "-", exp["amount"])
```

## Step 4: Function to Calculate Total Expense

```python
def calculate_total_expense():
    total = 0
    for exp in expenses:
        total = total + exp["amount"]
    return total
```

## Step 5: Function to Filter by Category

```python
def filter_by_category(c):
    filtered_list = []
    for exp in expenses:
        if exp["category"] == c:
            filtered_list.append(exp)
    return filtered_list
```

## Step 6: Function to Get Category Summary

```python
def get_category_summary():
    summary = {}
    for exp in expenses:
        c = exp["category"]
        a = exp["amount"]
        if c in summary:
            summary[c] = summary[c] + a
        else:
            summary[c] = a
    return summary
```

## Step 7: Function to Get the Highest Expense

```python
def get_highest_expense():
    high = expenses[0]
    for exp in expenses:
        if exp["amount"] > high["amount"]:
            hig = exp
    return high
```

## Step 8: Function to Display Expense Report Using Pandas

```python
def display_expense_report():
    df = pd.DataFrame(expenses)
    print(df)
    return df
```

## Step 9: Challenge Function - Category With Most Spending

```python
def get_top_spending_category():
    summary = get_category_summary()
    top_c = None
    top_a = 0
    for c in summary:
        if summary[c] > top_a:
            top_a = summary[c]
            top_c = c
    return top_c, top_a
```

## Step 10: Adding Sample Expenses

```python
add_expense("Food", 1200, "2026-07-01")
add_expense("Transport", 500, "2026-07-02")
add_expense("Shopping", 3500, "2026-07-03")
add_expense("Bills", 4500, "2026-07-04")
add_expense("Entertainment", 800, "2026-07-05")
add_expense("Food", 950, "2026-07-06")
```

## Step 11: Viewing All Expenses

```python
view_expenses()
```

## Step 12: Viewing Total Spending

```python
total_spd = calculate_total_expense()
print("Total Spending:", total_spd)
```

## Step 13: Viewing Spending by Category

```python
c_summary = get_category_summary()
print(c_summary)
```

## Step 14: Viewing the Most Expensive Transaction

```python
high_exp = get_highest_expense()
print("Highest Expense:", high_exp)
```

## Step 15: Displaying the Expense Report Using Pandas

```python
display_expense_report()
```

## Step 16: Showing the Category With Most Spending

```python
top_c, top_a = get_top_spending_category()
print("You spent the most on:", top_c, "with a total of", top_a)
```

## Step 17: Interactive Menu

```python
def run_menu():
    while True:
        print("1. Add expense")
        print("2. View all expenses")
        print("3. View total spending")
        print("4. View spending by category")
        print("5. View most expensive transaction")
        print("6. View expense report")
        print("7. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            c = input("Enter category: ")
            a = float(input("Enter amount: "))
            d = input("Enter date: ")
            add_expense(c, a, d)
        elif choice == "2":
            view_expenses()
        elif choice == "3":
            total_spd = calculate_total_expense()
            print("Total Spending:", total_spd)
        elif choice == "4":
            c_summary = get_category_summary()
            print(c_summary)
        elif choice == "5":
            high_exp = get_highest_expense()
            print("Highest Expense:", high_exp)
        elif choice == "6":
            display_expense_report()
        elif choice == "7":
            print("Exiting the program")
            break
        else:
            print("Invalid choice, please try again")
```
