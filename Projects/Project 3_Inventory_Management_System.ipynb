# Project 3: Simple Inventory Management System

## Import Required Libraries

```python
import pandas as pd
```

## Step 1: Create the Data Storage

```python
products = []
```

## Step 2: Function to Add a Product

```python
def add_product(n, p, q):
    pdt = {
        "name": n,
        "price": p,
        "quantity": q
    }
    products.append(pdt)
    print(n, "has been added to inventory")
```

## Step 3: Function to View All Products

```python
def view_products():
    for pdt in products:
        print(pdt["name"], "-", "Price:", pdt["price"], "-", "Quantity:", pdt["quantity"])
```

## Step 4: Function to Search a Product by Name

```python
def search_product(n):
    for pdt in products:
        if pdt["name"].lower() == n.lower():
            return pdt
    return None
```

## Step 5: Function to Update Product Quantity

```python
def update_quantity(n, new_q):
    pdt = search_product(n)
    if pdt is not None:
        pdt["quantity"] = new_q
        print(n, "quantity updated to", new_q)
    else:
        print(n, "not found in inventory")
```

## Step 6: Function to Calculate a Single Product's Value

```python
def calculate_product_value(pdt):
    return pdt["price"] * pdt["quantity"]
```

## Step 7: Function to Calculate Total Inventory Value

```python
def calculate_inventory_value():
    total_v = 0
    for pdt in products:
        total_v = total_v + calculate_product_value(pdt)
    return total_v
```

## Step 8: Function to Show Low Stock Products

```python
def show_low_stock_products():
    low_stock_list = []
    for pdt in products:
        if pdt["quantity"] < 5:
            low_stock_list.append(pdt)
    return low_stock_list
```

## Step 9: Function to Display Inventory Using Pandas

```python
def display_inventory():
    df = pd.DataFrame(products)
    print(df)
    return df
```

## Step 10: Adding Sample Products

```python
add_product("Rice Bag", 2500, 10)
add_product("Cooking Oil", 800, 3)
add_product("Sugar Bag", 1200, 15)
add_product("Tea Pack", 350, 4)
add_product("Soap", 100, 20)
```

## Step 11: Viewing All Products

```python
view_products()
```

## Step 12: Searching for a Product

```python
fnd_pdt = search_product("Tea Pack")
print(fnd_pdt)
```

## Step 13: Updating a Product Quantity

```python
update_quantity("Cooking Oil", 25)
```

## Step 14: Calculating Total Inventory Value

```python
total_v = calculate_inventory_value()
print("Total Inventory Value:", total_v)
```

## Step 15: Showing Low Stock Products

```python
low_stock_items = show_low_stock_products()
for i in low_stock_items:
    print(i["name"], "is low on stock with quantity", i["quantity"])
```

## Step 16: Displaying Full Inventory Using Pandas

```python
display_inventory()
```

## Step 17: Challenge - Menu Based System

```python
def run_menu():
    while True:
        print("1. Add product")
        print("2. View inventory")
        print("3. Search product")
        print("4. Update quantity")
        print("5. View low stock products")
        print("6. View total inventory value")
        print("7. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            n = input("Enter product name: ")
            p = float(input("Enter product price: "))
            q = int(input("Enter product quantity: "))
            add_product(n, p, q)
        elif choice == "2":
            display_inventory()
        elif choice == "3":
            n = input("Enter product name to search: ")
            fnd_pdt = search_product(n)
            print(fnd_pdt)
        elif choice == "4":
            n = input("Enter product name to update: ")
            new_q = int(input("Enter new quantity: "))
            update_quantity(n, new_q)
        elif choice == "5":
            low_stock_items = show_low_stock_products()
            for i in low_stock_items:
                print(i["name"], "is low on stock with quantity", i["quantity"])
        elif choice == "6":
            total_v = calculate_inventory_value()
            print("Total Inventory Value:", total_v)
        elif choice == "7":
            print("Exiting the program")
            break
        else:
            print("Invalid choice, please try again")
```
