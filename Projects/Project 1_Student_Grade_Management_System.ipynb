# Project 1: Student Grade Management System

## Import Required Libraries

```python
import pandas as pd
```

## Step 1: Create the Data Storage

```python
students = []
```

## Step 2: Function to Add a Student

```python
def add_student(n, m):
    std = {
        "name": n,
        "marks": m
    }
    students.append(std)
    print(n, "has been added")
```

## Step 3: Function to Calculate Total Marks

```python
def calculate_total(m):
    total = sum(m)
    return total
```

## Step 4: Function to Calculate Percentage

```python
def calculate_percentage(t, number_of_subjects):
    per = (t / (number_of_subjects * 100)) * 100
    return per
```

## Step 5: Function to Assign Grade

```python
def assign_grade(per):
    if per >= 90:
        grade = "A+"
    elif per >= 80:
        grade = "A"
    elif per >= 70:
        grade = "B"
    elif per >= 60:
        grade = "C"
    elif per >= 50:
        grade = "D"
    else:
        grade = "F"
    return grade
```

## Step 6: Function to Build the Full Student Record

```python
def process_student(std):
    total = calculate_total(std["marks"])
    per = calculate_percentage(total, len(std["marks"]))
    grade = assign_grade(per)
    std["total"] = total
    std["percentage"] = per
    std["grade"] = grade
    return std
```

## Step 7: Function to Display All Students Using Pandas

```python
def display_students():
    for std in students:
        process_student(std)
    df = pd.DataFrame(students)
    print(df)
    return df
```

## Step 8: Function to Get Class Average

```python
def get_class_average():
    for std in students:
        process_student(std)
    total_per = 0
    for std in students:
        total_per = total_per + std["percentage"]
    avg = total_per / len(students)
    return avg
```

## Step 9: Function to Get the Top Student

```python
def get_top_student():
    for std in students:
        process_student(std)
    top_std = students[0]
    for std in students:
        if std["percentage"] > top_std["percentage"]:
            top_std = std
    return top_std
```

## Step 10: Function to Get the Lowest Student

```python
def get_lowest_student():
    for std in students:
        process_student(std)
    lowest_std = students[0]
    for std in students:
        if std["percentage"] < lowest_std["percentage"]:
            lowest_std = std
    return lowest_std
```

## Step 11: Adding Sample Students

```python
add_student("Zuha Irfan", [85, 90, 78, 92, 88])
add_student("Sadaf Riaz", [45, 60, 55, 40, 50])
add_student("Rafia Fayyaz", [95, 98, 92, 97, 90])
add_student("Hajra Sarwar", [60, 40, 60, 45, 52])
```

## Step 12: Displaying All Students

```python
display_students()
```

## Step 13: Showing Class Average

```python
avg = get_class_average()
print("Class Average Percentage:", avg)
```

## Step 14: Showing the Top Student

```python
top_std = get_top_student()
print("Top Student:", top_std["name"], "with percentage", top_std["percentage"])
```

## Step 15: Showing the Lowest Student

```python
lowest_std = get_lowest_student()
print("Lowest Student:", lowest_std["name"], "with percentage", lowest_std["percentage"])
```

## Step 16: Challenge - Menu Based System

```python
def run_menu():
    while True:
        print("1. Add student")
        print("2. View all students")
        print("3. View class average")
        print("4. View top student")
        print("5. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            n = input("Enter student name: ")
            marks_input = input("Enter marks separated by commas: ")
            marks_list = marks_input.split(",")
            marks = []
            for mark in marks_list:
                marks.append(int(mark))
            add_student(n, marks)
        elif choice == "2":
            display_students()
        elif choice == "3":
            avg = get_class_average()
            print("Class Average Percentage:", avg)
        elif choice == "4":
            top_std = get_top_student()
            print("Top Student:", top_std["name"], "with percentage", top_std["percentage"])
        elif choice == "5":
            print("Exiting the program")
            break
        else:
            print("Invalid choice, please try again")
```

```python
run_menu()
```
