## Step 2: A Small Example DataFrame
I will use this small example to practice the basics before using the survey data.

```python
people = {
    'first': ['John', 'Jane', 'Sam'],
    'last': ['Doe', 'Smith', 'Brown'],
    'email': ['JohnDoe@email.com', 'JaneSmith@email.com', 'SamBrown@email.com']
}
people_df = pd.DataFrame(people)
```

## Step 3: Series vs DataFrame
A single column from a DataFrame is called a Series. A DataFrame is basically a container of Series that share the same index.

```python
people_df['email']
```

```python
type(people_df['email'])
```

```python
type(people_df)
```

## Step 4: Selecting a Single Column

```python
people_df['last']
```

## Step 5: Selecting Multiple Columns
To select more than one column, I pass a list of column names.

```python
people_df[['last', 'email']]
```

## Step 6: Viewing All Column Names

```python
people_df.columns
```

## Step 7: Selecting Rows with iloc
iloc lets me select rows and columns using their integer position.

```python
people_df.iloc[0]
```

```python
people_df.iloc[[0, 1]]
```

```python
people_df.iloc[[0, 1], 2]
```

## Step 8: Selecting Rows with loc
loc lets me select rows and columns using their labels instead of their position.

```python
people_df.loc[0]
```

```python
people_df.loc[[0, 1]]
```

```python
people_df.loc[[0, 1], ['email', 'last']]
```

```python
people_df.loc[0:1, 'last':'email']
```
