# Video 4: Filtering - Using Conditionals to Filter Rows and Columns

## Learning Objective
In this notebook I am learning how to filter rows in a DataFrame based on a condition, and how to
combine multiple conditions together using and, or, and not.

## Step 1: Import Pandas and Load the Data

```python
import pandas as pd
```

## Step 2: A Small Example DataFrame

```python
people = {
    'first': ['John', 'Jane', 'Sam'],
    'last': ['Doe', 'Smith', 'Doe'],
    'email': ['JohnDoe@email.com', 'JaneSmith@email.com', 'SamDoe@email.com']
}
people_df = pd.DataFrame(people)
```

## Step 3: Creating a Filter
A filter is just a Series of True and False values based on a condition.

```python
filt = (people_df['last'] == 'Doe')
```

```python
filt
```

## Step 4: Applying the Filter
I can pass this filter into the DataFrame to only get the rows where the condition is True.

```python
people_df[filt]
```

I can also use loc together with the filter.

```python
people_df.loc[filt]
```

```python
people_df.loc[filt, 'email']
```

## Step 5: Combining Conditions with And
I can combine two conditions using the & symbol.

```python
filt = (people_df['last'] == 'Doe') & (people_df['first'] == 'John')
people_df.loc[filt]
```

## Step 6: Combining Conditions with Or
I can combine two conditions using the | symbol.

```python
filt = (people_df['last'] == 'Doe') | (people_df['first'] == 'Jane')
people_df.loc[filt]
```

## Step 7: Using Not to Reverse a Filter
The ~ symbol flips True to False and False to True.

```python
people_df.loc[~filt]
```

## Step 8: Apply Filtering on the Survey Data
Now I will use these same filtering techniques on the real survey dataset.

### Filtering for High Salaries

```python
high_salary = (df['ConvertedComp'] > 70000)
df.loc[high_salary, ['Country', 'ConvertedComp']]
```

### Filtering for a List of Countries
The isin method lets me check if a value is inside a list of values.

```python
countries = ['United States', 'India', 'United Kingdom', 'Germany', 'Canada']
filt = df['Country'].isin(countries)
df.loc[filt, ['Country', 'ConvertedComp']]
```

### Filtering Text with str.contains
I can check if a text column contains a certain word, which is useful for columns like LanguageWorkedWith that store multiple values in one string.

```python
filt = df['LanguageWorkedWith'].str.contains('Python', na=False)
df.loc[filt, ['Country', 'LanguageWorkedWith']]
```

### Combining Filters on the Survey Data

```python
filt = (df['Country'] == 'India') & (df['ConvertedComp'] > 70000)
df.loc[filt, ['Country', 'ConvertedComp']]
```
