## Step 2: A Small Example DataFrame

```python
people = {
    'first': ['John', 'Jane', 'Sam'],
    'last': ['Doe', 'Smith', 'Brown'],
    'email': ['JohnDoe@email.com', 'JaneSmith@email.com', 'SamBrown@email.com']
}
people_df = pd.DataFrame(people)
```

## Step 3: Renaming All Columns at Once

```python
people_df.columns = ['first_name', 'last_name', 'email']
```

```python
people_df
```

## Step 4: Renaming Columns with a Loop

```python
people_df.columns = [x.upper() for x in people_df.columns]
```

```python
people_df
```

## Step 5: Renaming Specific Columns with rename
This is the safest way to rename columns because I only need to mention the ones I want to change.

```python
people_df.rename(columns={
    'FIRST_NAME': 'first',
    'LAST_NAME': 'last',
    'EMAIL': 'email'
}, inplace=True)
```

```python
people_df
```

## Step 6: Updating a Single Row

```python
people_df.loc[2]
```

```python
people_df.loc[2] = ['John', 'Smith', 'JohnSmith@email.com']
```

```python
people_df
```

## Step 7: Updating a Single Value

```python
people_df.loc[2, 'last'] = 'Doe'
```

```python
people_df
```

## Step 8: Updating Values that Match a Filter

```python
filt = (people_df['email'] == 'JohnSmith@email.com')
people_df.loc[filt, 'last'] = 'Smith'
```

```python
people_df
```

## Step 9: Updating a Whole Column with apply
The apply method runs a function on every value in a Series.

```python
people_df['email'].apply(len)
```

```python
def update_email(email):
    return email.upper()

people_df['email'] = people_df['email'].apply(update_email)
```

```python
people_df
```

I can also use apply with a lambda function instead of writing a full function.

```python
people_df['email'] = people_df['email'].apply(lambda x: x.lower())
```

```python
people_df
```

## Step 10: Using map and replace
map will change values that are not listed into missing values, while replace keeps values that are not listed the same.

```python
people_df['first'].map({'John': 'Chris', 'Jane': 'Mary'})
```

```python
people_df['first'] = people_df['first'].replace({'John': 'Chris', 'Jane': 'Mary'})
```

```python
people_df
```
