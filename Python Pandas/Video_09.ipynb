# Video 9: Cleaning Data - Casting Datatypes and Handling Missing Values

## Learning Objective
In this notebook I am learning how to find and handle missing values in a DataFrame, how to replace
custom missing value markers, and how to convert a column into the correct data type.

## Step 1: Import Pandas and NumPy

```python
import pandas as pd
import numpy as np
```

## Step 2: A Small Example DataFrame with Missing Values
I made this small example on purpose so it has different kinds of missing values, like NaN, None, and the text 'Missing'.

```python
people = {
    'first': ['John', 'Jane', 'Sam', np.nan, None, 'NA'],
    'last': ['Doe', 'Smith', np.nan, np.nan, 'Missing', 'Missing'],
    'email': ['JohnDoe@email.com', 'JaneSmith@email.com', 'SamBrown@email.com', np.nan, None, 'Anonymous@email.com'],
    'age': ['33', '55', '63', np.nan, None, 'Missing']
}
small_df = pd.DataFrame(people)
```

```python
small_df
```

## Step 3: Dropping Missing Values
By default, dropna removes any row that has at least one missing value.

```python
small_df.dropna()
```

I can be more specific about how dropna should work.

```python
small_df.dropna(axis='index', how='any')
```

```python
small_df.dropna(axis='index', how='all')
```

```python
small_df.dropna(axis='columns', how='any')
```

I can also drop rows only if a specific column has a missing value.

```python
small_df.dropna(axis='index', how='any', subset=['last'])
```

## Step 4: Replacing Custom Missing Value Markers
Right now, 'NA' and 'Missing' are just normal text, not real missing values. I need to convert them into proper NaN values first.

```python
small_df.replace('NA', np.nan, inplace=True)
```

```python
small_df.replace('Missing', np.nan, inplace=True)
```

```python
small_df
```

## Step 5: Checking for Missing Values with isna

```python
small_df.isna()
```

## Step 6: Filling Missing Values

```python
small_df.fillna('MISSING')
```

## Step 7: Casting Data Types
Right now, the age column is stored as text, even though it holds numbers.

```python
small_df.dtypes
```

```python
small_df['age'] = small_df['age'].astype(float)
```

```python
small_df.dtypes
```

Now that age is a number, I can run numeric functions like mean on it.

```python
small_df['age'].mean()
```

## Step 8: Apply These Skills on the Survey Data
The survey data uses custom text like 'NA' and 'Missing' for its missing values. I will load it again, but this time I will tell Pandas to treat these values as missing right away.

```python
na_vals = ['NA', 'Missing']
df = pd.read_csv('results.csv', na_values=na_vals)
```

## Step 9: Cleaning the YearsCode Column
The YearsCode column has some text answers mixed in with the numbers, like 'Less than 1 year' and 'More than 50 years'.

```python
df['YearsCode'].unique()
```

```python
df['YearsCode'] = df['YearsCode'].replace('Less than 1 year', 0)
```

```python
df['YearsCode'] = df['YearsCode'].replace('More than 50 years', 51)
```

## Step 10: Converting YearsCode to a Number

```python
df['YearsCode'] = df['YearsCode'].astype(float)
```

Now I can run numeric functions on this column.

```python
df['YearsCode'].mean()
```

```python
df['YearsCode'].median()
```
