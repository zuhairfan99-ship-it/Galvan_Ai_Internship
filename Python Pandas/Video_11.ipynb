# Video 11: Reading/Writing Data to Different Sources - Excel, JSON, SQL, Etc.

## Learning Objective
In this notebook I am learning how to read and write data using different file formats, including
CSV, Excel, JSON, and SQL.

## Step 1: Import Pandas and sqlite3

```python
import pandas as pd
import sqlite3
```

```python
pd.set_option('display.max_columns', 85)
pd.set_option('display.max_rows', 85)
```

## Step 2: Load the Survey Data

```python
df = pd.read_csv('results.csv')
```

## Step 3: Create a Smaller Sample of the Data
The full survey has almost 89,000 rows, so for this video I will just work with a small sample and a few columns to keep things fast and easy to look at.

```python
small_df = df[['Respondent', 'Country', 'ConvertedComp', 'LanguageWorkedWith']].head(50)
```

```python
small_df
```

## Step 4: Writing to a CSV File

```python
small_df.to_csv('survey_sample.csv', index=False)
```

## Step 5: Reading a CSV File Back In

```python
csv_df = pd.read_csv('survey_sample.csv')
```

```python
csv_df.head()
```

## Step 6: Writing to an Excel File

```python
small_df.to_excel('survey_sample.xlsx', index=False)
```

## Step 7: Reading an Excel File Back In

```python
excel_df = pd.read_excel('survey_sample.xlsx')
```

```python
excel_df.head()
```

## Step 8: Writing to a JSON File

```python
small_df.to_json('survey_sample.json', orient='records', lines=True)
```

## Step 9: Reading a JSON File Back In

```python
json_df = pd.read_json('survey_sample.json', orient='records', lines=True)
```

```python
json_df.head()
```

## Step 10: Writing to a SQL Database
I will use sqlite3 to create a simple local database and save my DataFrame into a table.

```python
conn = sqlite3.connect('survey_sample.db')
```

```python
small_df.to_sql('survey', conn, if_exists='replace', index=False)
```

## Step 11: Reading Data Back from SQL
I can use a normal SQL query to pull the data back out of the database and into a DataFrame.

```python
sql_df = pd.read_sql('SELECT * FROM survey', conn)
```

```python
sql_df.head()
```

## Step 12: Running a More Specific SQL Query

```python
high_earners = pd.read_sql('SELECT * FROM survey WHERE ConvertedComp > 50000', conn)
```

```python
high_earners
```

## Step 13: Closing the Database Connection

```python
conn.close()
```
