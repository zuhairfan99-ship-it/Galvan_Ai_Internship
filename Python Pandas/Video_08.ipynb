## Learning Objective
In this notebook I am learning how to summarize data using aggregate functions like median and mean,
how to count values, and how to group data by a column to analyze it in smaller pieces.

## Step 1: Import Pandas and Load the Data

```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', 85)
pd.set_option('display.max_rows', 85)
```

```python
df = pd.read_csv('results.csv')
```

## Step 2: Simple Aggregate Functions
Aggregate functions take many values and return a single value, like the median.

```python
df['ConvertedComp'].median()
```

I can use median on the whole DataFrame too, but I need to only include the numeric columns.

```python
df.median(numeric_only=True)
```

## Step 3: The describe Method
The describe method gives me many statistics about a column all at once.

```python
df['ConvertedComp'].describe()
```

## Step 4: Counting Values

```python
df['ConvertedComp'].count()
```

## Step 5: value_counts
The value_counts method counts how many times each value shows up in a column.

```python
df['Hobbyist'].value_counts()
```

I can also see these counts as percentages instead of raw numbers.

```python
df['SocialMedia'].value_counts(normalize=True)
```

## Step 6: Grouping Data with groupby
Grouping lets me split the data into smaller groups based on a column, so I can compare countries against each other.

```python
country_grp = df.groupby('Country')
```

I can pull out just one group, like all responses from the United States.

```python
country_grp.get_group('United States')
```

## Step 7: value_counts on a Group
Here I check what social media platform is most popular in India.

```python
country_grp['SocialMedia'].value_counts().loc['India']
```

## Step 8: Aggregate Functions on a Group

```python
country_grp['ConvertedComp'].median()
```

I can also run more than one aggregate function at the same time using agg.

```python
country_grp['ConvertedComp'].agg(['median', 'mean'])
```

## Step 9: Using apply with a Group
Here I count how many people in India know Python using a normal filter first.

```python
filt = df['Country'] == 'India'
df.loc[filt]['LanguageWorkedWith'].str.contains('Python', na=False).sum()
```

Now I will do the same thing for every country at once using groupby and apply.

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python', na=False).sum())
country_uses_python
```

## Step 10: Putting It All Together
Here I calculate the percentage of people in each country who know Python.

```python
country_respondents = df['Country'].value_counts()
```

```python
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'count': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
python_df
```

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython'] / python_df['NumRespondents']) * 100
```

```python
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
```
