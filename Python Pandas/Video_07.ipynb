and find the largest or smallest values in a column.

## Step 2: A Small Example DataFrame

```python
people = {
    'first': ['John', 'Jane', 'Sam', 'Adam'],
    'last': ['Doe', 'Smith', 'Brown', 'Doe'],
    'email': ['JohnDoe@email.com', 'JaneSmith@email.com', 'SamBrown@email.com', 'AdamDoe@email.com']
}
people_df = pd.DataFrame(people)
```

## Step 3: Sorting by One Column

```python
people_df.sort_values(by='last')
```

## Step 4: Sorting in Descending Order

```python
people_df.sort_values(by='last', ascending=False)
```

## Step 5: Sorting by Multiple Columns
Here I sort by last name first, and then by first name for people who share the same last name.

```python
people_df.sort_values(by=['last', 'first'], ascending=[False, True])
```

## Step 6: Sorting by the Index
After sorting the values, I can sort back to the original order using sort_index.

```python
people_df.sort_index()
```

## Step 7: Apply These Skills on the Survey Data

### Sorting Survey Data by Country

```python
df.sort_values(by='Country', inplace=True)
```

```python
df[['Country', 'ConvertedComp']].head()
```

### Sorting by Multiple Columns
I will sort by Country first, and then by salary from highest to lowest within each country.

```python
df.sort_values(by=['Country', 'ConvertedComp'], ascending=[True, False], inplace=True)
df[['Country', 'ConvertedComp']].head(20)
```

### Sorting Back to the Original Order

```python
df.sort_index(inplace=True)
```

## Step 8: Finding the Largest Values
The nlargest method is an easy way to find the rows with the highest values without sorting the whole DataFrame.

```python
df['ConvertedComp'].nlargest(10)
```

```python
df.nlargest(10, 'ConvertedComp')
```

## Step 9: Finding the Smallest Values

```python
df.nsmallest(10, 'ConvertedComp')
```
