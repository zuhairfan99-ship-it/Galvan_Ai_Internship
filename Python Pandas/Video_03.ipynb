# Video 3: Indexes - How to Set, Reset, and Use Indexes

## Learning Objective
In this notebook I am learning how to set a column as the index of a DataFrame, how to reset the
index back to normal, and how indexes make it easier to look up rows.

## Step 1: Import Pandas and Load the Data

```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', 85)
pd.set_option('display.max_rows', 85)
```

```python
df = pd.read_csv('results.csv')
```

```python
schema_df = pd.read_csv('schema.csv')
```

## Step 2: A Small Example DataFrame

```python
people = {
    'first': ['John', 'Jane', 'Sam'],
    'last': ['Doe', 'Smith', 'Brown'],
    'email': ['JohnDoe@email.com', 'JaneSmith@email.com', 'SamBrown@email.com']
}
people_df = pd.DataFrame(people)
```

## Step 3: Looking at the Default Index
By default, Pandas gives every DataFrame a numeric index that starts at 0.

```python
people_df
```

```python
people_df.index
```

## Step 4: Setting a Column as the Index
I can set the email column as the index since it is a unique value for every row.

```python
people_df.set_index('email')
```

Notice that the change was not saved because I did not use inplace or reassign the DataFrame.

```python
people_df
```

This time I will use inplace so the change is actually saved.

```python
people_df.set_index('email', inplace=True)
```

```python
people_df
```

```python
people_df.index
```

## Step 5: Using loc with the New Index
Now that email is the index, I can use loc to look up a row using the email value.

```python
people_df.loc['JaneSmith@email.com']
```

```python
people_df.loc['JaneSmith@email.com', 'last']
```

## Step 6: Resetting the Index
If I want to go back to the default numeric index, I can use reset_index.

```python
people_df.reset_index(inplace=True)
```

```python
people_df
```

## Step 7: Setting the Index While Loading the Data
I can also set the index directly when I load a csv file using the index_col argument.

```python
df = pd.read_csv('results.csv', index_col='Respondent')
```

```python
df.head()
```

## Step 8: Setting an Index on the Schema DataFrame
The schema file also has a natural index, which is the Column name.

```python
schema_df.set_index('Column', inplace=True)
```

```python
schema_df.head()
```

## Step 9: Looking Up a Question by Column Name
Now I can quickly look up the full question for any column name.

```python
schema_df.loc['Hobbyist']
```

```python
schema_df.loc['Hobbyist', 'QuestionText']
```

## Step 10: Sorting the Index
I can sort the schema DataFrame alphabetically by its index.

```python
schema_df.sort_index()
```

```python
schema_df.sort_index(ascending=False)
```
