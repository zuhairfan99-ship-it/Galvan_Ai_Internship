# Video 1: Getting Started with Data Analysis

## Learning Objective
In this notebook I am learning how to install Pandas, create a simple DataFrame, and load a real
dataset (the Stack Overflow Developer Survey) into a Jupyter Notebook. I will also learn how to look
at basic information about the data I loaded.

## Step 1: Import Pandas
First I need to import the pandas library.

```python
import pandas as pd
```

## Step 2: Create a Small Example DataFrame
Before working with the real dataset, I will practice by creating a small DataFrame from a Python dictionary.

```python
people = {
    'first': ['John', 'Jane', 'Sam'],
    'last': ['Doe', 'Smith', 'Brown'],
    'email': ['JohnDoe@email.com', 'JaneSmith@email.com', 'SamBrown@email.com']
}
```

Now I will turn this dictionary into a DataFrame.

```python
people_df = pd.DataFrame(people)
```

Let's look at this small DataFrame.

```python
people_df
```

## Step 3: Load the Real Dataset
Now I will load the Stack Overflow Developer Survey dataset that we will be using throughout this course.

```python
df = pd.read_csv('results.csv')
```

I also have a schema file that explains what each column means.

```python
schema_df = pd.read_csv('schema.csv')
```

## Step 4: Set Display Options
The dataset has a lot of columns, so I will change the display settings so I can see more columns and rows at once.

```python
pd.set_option('display.max_columns', 85)
pd.set_option('display.max_rows', 85)
```

## Step 5: View the Shape of the Data
The shape tells me how many rows and columns are in the dataset.

```python
df.shape
```

## Step 6: View Basic Information
The info method shows me the column names, how many non-null values each column has, and the data type of each column.

```python
df.info()
```

## Step 7: View the First Few Rows
